### Color to Gray

```python
import numpy as np

def rgb2gray(rgb):

    r, g, b = rgb[:,:,0], rgb[:,:,1], rgb[:,:,2]
    gray = 0.2989 * r + 0.5870 * g + 0.1140 * b

    return gray
```

```python
img=rgb2gray(img)
```

```python
img.shape
```

```python
imgplot = plt.imshow(img,cmap='gray')
```

### Define a Gaussian Function

The Gaussian function can be applied to a vector.

```python
from matplotlib import pyplot as plt
import numpy as np
import math
def gaussian(x, mu,sigma):
    #return np.exp(-sigma*np.power((x - mu), 2.))
    return np.exp(-np.power((x-mu),2.)/(2*sigma*sigma))
```

```python
x=[100,190,200,210,300]
x=np.asarray(x)
```

```python
gaussian(x,200,10)
```

```python
plt.plot(gaussian(x,200,20))
plt.title(r"Gaussian window ($\sigma$=10)")
plt.ylabel("Amplitude")
plt.xlabel("Sample")
```
