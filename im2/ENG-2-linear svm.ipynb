```python
from sklearn.svm import SVR
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
```

```python
img=mpimg.imread('fig1.png')
```

```python
imgplot = plt.imshow(img)
```

```python
import numpy as np

def rgb2gray(rgb):

    r, g, b = rgb[:,:,0], rgb[:,:,1], rgb[:,:,2]
    gray = 0.2989 * r + 0.5870 * g + 0.1140 * b

    return gray
```

```python
img=rgb2gray(img)
```

```python
img.shape
```

```python
imgplot = plt.imshow(img,cmap='gray')
```

```python
img.shape
```

```python
X=[]
Y=[]
```

```python
counter=0
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        if img[i][j]<0.7:
            img[i][j]=0
            Y.append(394-i)
            X.append(j/1000)
        else:
            img[i][j]=1
```

```python
imgplot = plt.imshow(img,cmap='gray')
```

```python
plt.plot(X,Y, 'b.')  # plot x and y using blue circle markers
```

```python
clf = SVR(kernel='rbf', C=1e5, epsilon=1e-7,gamma=1e5)
```

```python
X=np.asarray(X)
Y=np.asarray(Y)
```

```python
X=X.reshape(X.shape[0],1)
```

```python
clf.fit(X,Y)
```

```python
x=np.linspace(0., 1., 200)
```

```python
x=x.reshape(x.shape[0],1)
```

```python
Z=clf.predict(x)
```

```python
plt.plot(x,Z, 'b-')
```
